```python
ls = [1,2,3,4]
ls.append(5)
ls.append(5)
ls.append(5)
ls.append(5)
print(ls)
print(id(ls))
ls2 = ls.copy()  #复制一份新的
print(ls2)
print(id(ls2))
print(ls2.count(5))
print()
ls3 = ls2.extend([6,7,8,9])  # 在原来的列表上添加另一个列表内的元素
print()

print(ls2)
print(id(ls2))
print(ls3)   #extend 返回的None
print(id(ls3))  # None 的地址
print()

print(ls2.index(6,0,100))  # index()：返回序号
ls2.insert(1,10)
print(ls2)
a = ls2.pop() # 弹出最后一个元素
print(a)
print(ls2)
ls2.remove(8) # 移除第一个元素
print(ls2)
ls2.reverse()
print(ls2)

#sort（）: 排序 默认从小到大
ls2.sort()
print(ls2)

ls2.sort(reverse = True)  # 从大到小  # * in place* 原地排序
print(ls2)

ls.clear()
print(ls)
```
