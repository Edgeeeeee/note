# 信道的极限容量
- 从概念上将，限制马元在信道的传输速率的而因素有以下两个
    - 信道能狗通过的频率范围
    - 信噪比
- 在任何信道中，马元的传输速率是有上限的，否则就会数显骂间串扰的问题，世界手段对马原的判决（即识别）成为不可能
- 如果信道的频带越宽，也就是能狗通过的信号高频分量越多，那么就可以使用更高的速率传送码元而不出现码间串扰
- 噪声存在于所哟有的电子设备和通信信道中
- 噪声是随机产生的，他的瞬时值有时会很大。因此噪声会使接收端对码元的判决产生错误。
- 但噪声的影响是相对的，如果信号相对较强，那么噪声的影响就相对较小
- __信噪比__就是信号的平均功率和噪声的平均功率之比。
- 常常记为S/N，并用分贝作为度量的单位即
               信噪比 = 10 lg（S/N）
- 例如当S/N = 10 时，信噪比为10 dB ，而当S/N = 1000 时，信噪比为30dB
- 1984年，香浓用信息论的理论推导出了带宽受限且有高斯白噪声干扰的信道极限，无差错的信息传输速率
- 信道的极限信息传输速率C 课=可以表达为:
           C = W log(1 + S/N)
           其中：w为信道的带宽  s为信道内所传信号的平均功率 n为信道内部的高斯噪声功率
- 信道的带宽比越大，则信息的极限传输速率越高。
- 知道信息传输速率低于信道的极限信息传输速率，就一定可以找到某种办法来实现无差错的传输。
- 若信道带宽w或信噪比s/n没有上限，则信道的极限信息传输速率c也没有上限
- 实际信道上能够达到的信息传输速率要比香农的极限传输效率低不少
- 用编码的方法让码元携带更多比特的信息量

# 物理层下面的传输媒体
- 导引型传输媒体
- 非导引型传输媒体

- 传输媒体也称为，传输介质或者传输媒介，他就是数据传输系统中在发送器和接收器之间的物理通路
- 传输媒体课分为两大类，即导引型传输媒体和非导引型传输媒体
- 在导引型传输媒体中，电磁波被导引沿着固体媒体传播
- 非导引型传输媒体就是指自由空间。非导引型传输媒体中，电磁波的传输通常称为无线传输
- 当光线从高折射率的媒体射向低折射率的媒体时，其折射角将大于入射角，如果入射角足够大，就会出现全反射，光也就沿着光纤传输下去

#### 非导引型传输媒体

- 将自由空间称为非导引型传输媒体。
- 五i先传输所使用的频段很光
- __短波通信__主要是靠电离层的反射，但短波信道的通信质量较差，传输速率低
- __微波__在空间主要是直线传播
- 传统微博通信有两种方式
    - 地面微波接力通信
    - 卫星通信

# 信道复用技术
- 频分复用，时分复用和统计时分复用
- 波分复用
- 码分复用

复用是通信技术中的基本概念，
他允许用户使用一个**共享**i信道进行通信，降低成本，提高利用率

### 频分复用
- 将整个带宽分为多份，用户再分配到一定的频带后，在通信过程始终都占用这个频带
- 频分复用的所有用户在同样的时间占用不同的带宽资源
### 时分复用
- 将实际俺划分为一段段登场的时分复用帧（TDM帧），每一个时分复用的用户在每一个tdm帧中站有固定序号的时隙
- 每一个用户占有的时隙是周期性出现的
- tdm信号也成为__等时__信号
- 时分复用的所有用户是在不同时间占用同样的频带宽度
- 使用时分复用系统传送计算机数据时，由于计算机数据的突发性，用户对分配到的子信道的利用率一般是不高的，会造成线路资源浪费的现象
- 统计时分复用是 不固定的分配时隙，而是按需动态的分配时隙，因此统计时分复用可以提高线路的利用率

### 波分复用


就是光的频分复用。是用一根光纤来同时传送索格光载波信号。

### 码分复用
- 常用的名词是码分多址
- 个用户使用经过特殊特殊挑选的不同码型，因此彼此不会造成干扰
- 这种系统发送的信号有很强的抗干扰能力，其频谱类似于白噪声，不易被敌人发现

- 码片序列： 每一个比特时间划分为m个短的间隔，称为码片
- 每个站被指派唯一的 m bit 码片序列
    - 如发送比特1，则发送自己的 m 比特码片序列
    - 如果发送比特0，则发送该码片序列的二进制反码
- 向量规格化是指 1 用 1 表示，0 用 -1 表示

- 这种通信方式是扩频通信中的一种
- 扩频通信通常分为两大类
    - 一种是直接序列扩频DSSS
    - 另一种是调频扩频FHSS
- 每个站分配的码片序列不仅必须各不相同，并且嗨必须相互正交
- 在使用的系统中是使用伪随机码序列
- 两个不同站的码片序列内积为0
- 任何一个码片向量和该码片向量自己的规格化内积都是1
- 一个码片向量和该码片向量反码的向量的规格化内积值是-1

# 数字传输系统
- 与模拟通信相比，数字通信无论是在传输质量还是经济上都有明显的又是
- 目前长途干线大都采用时分复用PCM的数字传输方式
- 脉码调制 PCM 体制最初是为了在电话局之间的中继线上传送多路的电话

# 旧的数字传输系统存在许多缺点
- 速率标准不统一
    - 如果不对高次群的数字传输速率进行标准化，国际范围的基于光线高速数据传输就很难实现
- 不是同步传输
    - 在过去相当时间里，为了节约经费，各国的数字网是采用准同步方式
    - 当数据传输的速率很高时，收发双方的始终同步就成为很大的问题

# 宽带接入技术
- 美国联邦通信委员会FCC认为只要双向速率之和超过 200 kbit/s 就是宽带
- 从宽带接入的媒体来看可划分为两大类
    - 有线宽带接入
    - 无线宽带接入
- ADSL技术
    - 非对称数字用户线ADSL技术就是用数字技术对现有的模拟电话用户线进行改造，使它能够承载宽带业务
    - adsl技术把0 - 4 khz 地段频谱留给传统电话使用，把原来没有利用的高频谱留给用户上网使用
    - DSL就是数字用户线的缩写

# ADSL的特点
- 上行和下行带宽做成不对称的
    - 上行值用户到isp，下行只从isp到用户
- ADSL在用户先的端各安装一个ADSL调制解调器
- 我目前采用的方案是离散多音调 DMT
    - 这里的多音调 就是 多载波或者多子信道 的意思
- DMT调制技术采用频分复用的额方法，把40 kHz以上一直到1.1 MHz的额高端频谱划分为许多的子信道，其中25个用于上行信道，249个子信道用于下行信道
- 每个子信道占据4kHz带宽，并且使用不同的载波进行数字调制。这种做法相当于在一对用户线上使用许多小的调制解调器并行的传送数据

# ADSL的数据率
- ADSL采用自适应调至即使是用户线能够传送尽可能高的数据率
- ADSL不能保证固定的数据率
- ADSL的组成
    - DSLAM: 数字用户接入复用器
    - ATU: 接入端单元
    - ATU-C  ATU-R
    - PS: 电话分离器

# 第二代 ADSL
- 通过提高调至效率得到了更高的数据率
- 采用了无缝速率自适应技术SRA，可以在运营中不中断通信和不产生误码的情况下自适应的调整数据

# HFC 网对 CATV 网进行了改造
- HFC网江原CATV网中的同轴电缆主干部分改换位光纤，并使用模拟光纤技术
- 在模拟光纤中采用光的振幅调制AM ，这比使用数字光线更为经济
- 莫尼冈县从**头端**连接到**光纤结点** ，即**光分配结点ODN**。在光线结点光信号被转换为电信号。在管线节点一下就是同轴电缆

- 电缆调制解调器是为HFC而使用的调制解调器
